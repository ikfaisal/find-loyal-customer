```python
# Convert to DateTime

df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])
df['order_delivered_carrier_date'] = pd.to_datetime(df['order_delivered_carrier_date'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
df['shipping_limit_date'] = pd.to_datetime(df['shipping_limit_date'])
```

```python
def sns_barplot(df, x, y, xLabel, yLabel, title, orient='v'):
    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(10, 8))

    #plot the missing value count
    sns.set(color_codes=True)
    sns.barplot(x=x, y=y, data=df, orient='v')

    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.title(title)

    plt.xticks(rotation = 90)

    plt.show();
```

```python
# Find out in which state most customers are coming from

customerCountDataFrame = pd.DataFrame((df.groupby(by=['customer_state'])['customer_id'].count() / df.shape[0] * 100).reset_index())
```

```python
# Most customers are coming from SP, RJ, MG

sns_barplot(customerCountDataFrame.sort_values(by='customer_id', ascending=False), 'customer_state', 'customer_id', 'State', 'Number of Customer', 'Statewise Customer')
```

```python
# Most revenue generated state

totalPriceDataFrame = pd.DataFrame((df.groupby(by=['customer_state'])['price'].sum()).reset_index())
totalPriceDataFrame
```

```python
# Total revenues are coming from SP, RJ, MG

totalPriceDataFrame['price'] = totalPriceDataFrame['price'] / totalPriceDataFrame['price'].sum() * 100
sns_barplot(totalPriceDataFrame.sort_values(by='price', ascending=False), 'customer_state', 'price', 'State', 'Total Price (%)', 'Statewise Total Revenue');
```

```python
# Most revenue generated category

totalCategoryDataFrame = pd.DataFrame((df.groupby(by=['product_category_name_english'])['price'].sum()).reset_index())
totalCategoryDataFrame
```

```python
# Total revenues are coming from SP, RJ, MG

# totalPriceDataFrame['price'] = totalPriceDataFrame['price'] / totalPriceDataFrame['price'].sum() * 100
sns_barplot(totalCategoryDataFrame.sort_values(by='price', ascending=False).head(10), 'product_category_name_english', 'price', 'Category', 'Total Revenue', 'Productwise Revenue');
```

```python
# creating an aggregation
avg_score_per_category = df.groupby('product_category_name_english', as_index=False).agg({'review_score': ['count', 'mean']})
avg_score_per_category.columns = ['product_category_name_english', 'count', 'mean']

# filtering to show only categories with more than 50 reviews
avg_score_per_category = avg_score_per_category[avg_score_per_category['count'] > 50]
avg_score_per_category = avg_score_per_category.sort_values(by='mean', ascending=False)
avg_score_per_category
```

```python
# Most revenue generated category

sns_barplot(avg_score_per_category.sort_values(by='mean', ascending=False), 'product_category_name_english', 'mean', 'Category', 'Score', 'Product Review', orient='h');
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 8))

#plot the missing value count
sns.set(color_codes=True)
sns.barplot(x='mean', y='product_category_name_english', data=avg_score_per_category.head(15), orient='h')

plt.xlabel('Review Score')
plt.ylabel('Product')
plt.title('Popular Product')

plt.show();
```

```python
import datetime
deliveryDelayDataFrame = df[df.order_delivered_customer_date - df.order_estimated_delivery_date > datetime.timedelta(days=0)]
```

```python
deliveryDelayDataFrame
```

```python
# creating an aggregation
avg_score_per_category_delay = deliveryDelayDataFrame.groupby('product_category_name_english', as_index=False).agg({'review_score': ['count', 'mean']})
avg_score_per_category_delay.columns = ['product_category_name_english', 'count', 'mean']

# filtering to show only categories with more than 50 reviews
avg_score_per_category_delay = avg_score_per_category_delay[avg_score_per_category_delay['count'] > 50]
avg_score_per_category_delay = avg_score_per_category_delay.sort_values(by='mean', ascending=False)
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 8))

#plot the missing value count
sns.set(color_codes=True)
sns.barplot(x='mean', y='product_category_name_english', data=avg_score_per_category_delay.head(10), orient='h')

plt.xlabel('Review Score')
plt.ylabel('Product')
plt.title('Popular Product')

plt.show();
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25

# set height of bar
bars1 = merged_dataframe['product_category_name_english']
bars2 = merged_dataframe['mean_x']
bars3 = merged_dataframe['mean_y']

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Make the plot
# plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='var1')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='Review ')
plt.bar(r3, bars3, color='r', width=barWidth, edgecolor='white', label='Review (Delay)')

# Add xticks on the middle of the group bars
plt.xlabel('Product Review', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], [p for p in bars1], rotation = 90)

# Create legend & Show graphic
plt.legend()
plt.show()
```
